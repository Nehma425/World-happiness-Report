```python
# Importing cleaned individual years as CSV files 
df_2015 =pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'cleaned_2015.csv'), index_col = False)
df_2016 =pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'cleaned_2016.csv'), index_col = False)
df_2017 = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'cleaned_2017.csv'), index_col = False)
df_2018 =pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'cleaned_2018.csv'), index_col = False)
df_2019 = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'cleaned_2019.csv'), index_col = False)
```

```python
# Joining 2015 and 2016
df_2015_2016 = pd.merge(df_2015, df_2016, how='outer', on=['Country', 'Region', 'Year', 'Happiness_rank', 'Happiness_score', 'GDP_capita', 'Family', 'Life_expectancy', 'Freedom', 'Gov_trust', 'Generosity', 'Dystopia_residual'])
```

```python
df_2015_2016.tail()
```

```python
df_2015_2016.shape
```

```python
# Joining (2015-2016) with 2017
df_15_16_17 = pd.merge(df_2015_2016, df_2017, how='outer', on=['Country', 'Region', 'Year', 'Happiness_rank', 'Happiness_score', 'GDP_capita', 'Family', 'Life_expectancy', 'Freedom', 'Gov_trust', 'Generosity', 'Dystopia_residual'])
```

```python
df_15_16_17.head()
```

```python
df_15_16_17.shape
```

```python
#now with 2018
# Joining (15-17) with 2018
df_15_16_17_18 = pd.merge(df_15_16_17, df_2018, how='outer', on=['Country', 'Region', 'Year', 'Happiness_rank', 'Happiness_score', 'GDP_capita', 'Family', 'Life_expectancy', 'Freedom', 'Gov_trust', 'Generosity'])
```

```python
df_15_16_17_18.shape
```

```python
# Joining (15-18) with 2019
merge_all_years = pd.merge(df_15_16_17_18, df_2019, how='outer', on=['Country', 'Region', 'Year', 'Happiness_rank', 'Happiness_score', 'GDP_capita', 'Family', 'Life_expectancy', 'Freedom', 'Gov_trust', 'Generosity'])
```

# 3. Consistency Checks

```python
# Checking for missing values
merge_all_years.isnull().sum()  #one missing value for Gov_trust and 312 for dustopia_residual(due to the fact that it was not in original data)
```

```python
# Checking for duplicates
merge_all_years.dups = merge_all_years[merge_all_years.duplicated()] 
```
